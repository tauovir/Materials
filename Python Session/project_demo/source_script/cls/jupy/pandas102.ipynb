The **iloc[]** operator works with the same rules as the selection operator[] for list in python.<br>
It, focuses on the position of the elements expressed by sequences of integers from 0 to N-1.<br>
**iloc[row number, column number]**


```python
# data1.loc['b','city'] = data.iloc[1,3]
# data1.loc['c','state'] =data.iloc[2,4]
```

## The Boolean Conditioan


```python
dict_list = {"id" : [101,102,103,104,105,106],
            "name" : ["Orange","Apple","Mango","Banana","date","pineapple"],
             "price" : [120,150,160,180,200,300],
             "city" : ["meerut","kanpur","mumbai","goa","delhi","indore"],
             'state' : ["Uttar Pradesh","Madhya Pradesh","Maharashtra","Kerala","Gujrat","Asam"],
             'country':["india","US","US","india","US","india"]
            }
data1 = pd.DataFrame(dict_list)
data1
```

```python
data1['price'] >160
```

```python
data1[data1['price'] > 160]  # Boolean index
```

```python
data1['country'] == 'US'
```

```python
data1[data1['country'] =='US']  # Boolean index
```

```python
data1[(data1['country'] =='india') & (data1['price'] > 160)]  # &(and), |(or),~(not)
```

## Deleting Rows and Columns from DataFrame
We can use DataFrame.drop() method to delete rows and columns from DataFrame.<br>
We need to specify the name of labels to be dropped and axis from which they need to be dropped<br>
To delete a row, parameter axis assigned to value 0,<br>
To delete a columns ,parameter axis assigned to value 1

```python
data2 = data1.copy()
data2
```

```python
data2 = data.drop(2, axis = 0)
data2
```
