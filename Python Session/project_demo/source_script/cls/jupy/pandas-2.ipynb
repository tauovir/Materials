```python
def convertSqftToNum(sqft):
  
    
    list1 = sqft.split("-")
    if len(list1) == 2:
        x1 = float(list1[0])
        x2 = float(list1[1])
        return (x1 + x2) / 2
    try:
        return float(list1[0])
    except : 
        return None
    
```

```python
nadata['t_sqft'] = nadata['total_sqft'].apply(convertSqftToNum) ## Call custome function
nadata
```
