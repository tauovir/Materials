## Fill missing values


```python
# Check method
#Filling null values with a single value
# data.fillna('AAAA')
nadata = data.copy()
```

```python
nadata['society'] = nadata['society'].fillna('Cantt')
nadata['balcony'].fillna(nadata['balcony'].max(), inplace =True)
```

```python
nadata = data.copy()
# Filling null values with the previous ones
#Filling null value with the next ones
```

```python
# Filling null values with the previous ones
# nadata['society'] = nadata['society'].fillna(method = 'pad')
#Filling null value with the next ones
nadata['society'] = nadata['society'].fillna(method = 'backfill')
```

```python
nadata['society'] = nadata['society'].fillna(method = 'ffill')
nadata['balcony'] = nadata['balcony'].fillna(nadata['balcony'].max())
```
