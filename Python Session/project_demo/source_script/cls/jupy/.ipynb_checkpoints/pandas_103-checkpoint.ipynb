# Add data to dataFrame
**append**: is suitable for adding new data as new rows of starting Dataframe<br>
**concate**:is suitable for adding new data as new columns of DataFrame <br>

```python
dict_list = {"id" : [101,102,103],
            "name" : ["Orange","Apple","Mango"],
             "price" : [120,150,160],
             "city" : ["meerut","kanpur","mumbai"],
            }
main_data = pd.DataFrame(dict_list)
data = main_data.copy()
data
```

```python
dict2 = dict_list = {"id" : [201,202,203,],
             "name" : ["Banana","date","pineapple"],
             "price" : [180,200,300],
             "city" : ["goa","delhi","indore"],
                "country" : "india"
            }
data2 = pd.DataFrame(dict2)
data2
```

### append
The append method called on a DataFrame allows the addition of the data of second DataFrame starting from the last row. If the two data frame have consistent strunture , we ill have same result

```python
df3 = data.append(data2)
df3
```

```python
df3 = data.append(data2,ignore_index = True)
df3
```

```python
user = pd.read_csv('data/user101.csv')
address = pd.read_csv('data/addr3.csv')
```

```python
user
```

```python
address
# address.drop(['id'],axis=1,inplace = True)
```

```python
address.drop([3,4,5], inplace=True)
# 
```

```python
pd.merge(user, address, left_on = 'id',right_on = "user_id")
```

```python
#only include observation found in both dataframe
pd.merge(user,address, left_on = 'id',right_on = "user_id", how="inner")
```

```python
#outer: include obsrvation found in either A or B
#left:include all observation from left(user)
#left:include all observation from right(address)
#
pd.merge(user,address, left_on = ['id'],right_on = ["user_id"], how="left")
```

```python
pd.merge(user,address, left_on = ['id'],right_on = ["user_id"], how="right")
```

```python
pd.merge(user,address, left_on = ['id'],right_on = ["user_id"], how="outer")
```

```python
## Connect with Postgres SQL

import sqlalchemy
from sqlalchemy.exc import SQLAlchemyError
DB_USER = "postgres"
DB_PASSWORD = "khan102030"
DB_NAME = "my_database"
DB_HOST = "localhost"
DB_PORT = 5432
def get_engine():
    try:
        #engine = create_engine('postgresql+psycopg2://scott:tiger@localhost/mydatabase')
        connection_url = f"postgresql+psycopg2://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
        engine = sqlalchemy.create_engine(connection_url)
        return engine
    except SQLAlchemyError as e:
        print("Exception")
        raise (e)

def _get_query():
    user_ql = "select * from users"
    add_sql = "select * from address"
    inner_sql  = """
                select ad.addr,ad.city,country,
                u.id as user_id,u.name,u.mobile,u.salary,u.created_at from users u
                inner join address ad on ad.user_id = u.id
    
                """
    left_sql  = """
                select ad.addr,ad.city,country,
                u.id as user_id,u.name,u.mobile,u.salary,u.created_at from users u
                left join address ad on ad.user_id = u.id
    
                """
    
    return [user_ql,add_sql,inner_sql,left_sql]
```
