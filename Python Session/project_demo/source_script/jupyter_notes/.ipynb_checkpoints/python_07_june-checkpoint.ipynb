## Boolean condition

```python
dict_list = {
    "id" : [101,102,103,104,105,106],
    "name" : ["apple","orange","mango","date","banana","pineapple"],
    "price" : [120,130,140,150,160,170],
    "city" :["meerut","kanpur","mumbai","delhi","indore","goa"],
    "state" :["up","mp","maharashtra","kerala","gujrat","asam"],
    "country" : ["india","US","US","india","india","US"]
}
data = pd.DataFrame(dict_list)
data
```

```python
data['price'] > 140
```

```python
data[data['price'] > 150]
```

```python
data['country'] == 'US'
```

```python
data[data['country'] == 'US']
```

```python
data[(data['price'] > 150) & (data['country'] == 'US')]
```

```python
data.head(10)
```
