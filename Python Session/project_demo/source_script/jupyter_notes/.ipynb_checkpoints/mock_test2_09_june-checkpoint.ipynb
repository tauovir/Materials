7. WAP to read a string fron keyboard,and create dictioinary containing frequence of each character
occuring in the string.ex,str = "apple", dict1 = {'a':1,'p':2,'l':1,'e':1}

```python
name = input("Enter String:")
dict1 = {}
for ch in name:
    if ch in dict1:
        dict1[ch] = dict1[ch] + 1
    else:
        dict1[ch] =  1
        
dict1       
```
