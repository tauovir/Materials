```python
import sqlalchemy
from sqlalchemy.exc import SQLAlchemyError
DB_USER = "postgres"
DB_PASSWORD = "khan102030"
DB_NAME = "my_database"
DB_HOST = "localhost"
DB_PORT = 5432


def get_engine():
    try:
        #engine = create_engine('postgresql+psycopg2://scott:tiger@localhost/mydatabase')
        connection_url = f"postgresql+psycopg2://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
        engine = sqlalchemy.create_engine(connection_url)
        return engine
    except SQLAlchemyError as e:
        print("Exception")
        raise (e)


def get_db_query():
    user_sql = "SELECT * FROM users ORDER BY id ASC "
    join_query = """
    SELECT ad.addr, ad.city,ad.country,
    u.id as user_id,u.name,u.mobile,u.salary,u.created_at
    FROM users u
    left join address ad on ad.user_id = u.id;

    """
    
    return [user_sql,join_query]
```
